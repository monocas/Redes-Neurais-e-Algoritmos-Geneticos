## Código e discussão



### Relembrando o que temos



No notebook anterior construímos a seguinte classe chamada `Valor`:



```python
class Valor:
    def __init__(self, data, progenitor=(), operador_mae="", rotulo=""):
        self.data = data
        self.progenitor = progenitor
        self.operador_mae = operador_mae
        self.rotulo = rotulo

    def __repr__(self):
        return f"Valor(data={self.data})"

    def __add__(self, outro_valor):
        data = self.data + outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "+"
        saida = Valor(data, progenitor, operador_mae)
        return saida

    def __mul__(self, outro_valor):
        data = self.data * outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "*"
        saida = Valor(data, progenitor, operador_mae)
        return saida
```

Com essa classe nós conseguimos plotar o grafo computacional que criamos na primeira aula sobre redes neurais artificiais usando a função `plota_grafo` criada por Andrej. Esse grafo computacional representa uma rede neural que nós criamos para prever o valor $y$.



```python
x1 = Valor(60, rotulo="x1")
x2 = Valor(24, rotulo="x2")
w1 = Valor(10, rotulo="w1")
w2 = Valor(5, rotulo="w2")
w3 = Valor(2, rotulo="w3")
b = Valor(7, rotulo="b")

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

plota_grafo(y)
```

### Preparando o terreno para o cálculo do gradiente local



Cada valor numérico do grafo computacional deve ter um valor de `gradiente local` associado a ele. O gradiente local representa a derivada parcial do vértice final do grafo computacional (que é $y$ no nosso caso, também conhecido como vértice folha) em relação a o valor do vértice em questão.

Suponha que nosso vértice folha se chame $y$. Agora suponha que nós temos um vértice qualquer no nosso grafo computacional que se chama $w_2$. O gradiente local de $w_2$ nada mais é do que:

$$
\frac{\partial y}{\partial w_2}
$$

Se cada valor tem um gradiente local, então precisamos registrar isso na nossa classe. Vamos criar uma propriedade chamada `grad`. Vamos dar o valor padrão de zero para essa propriedade. Na próxima seção iremos alterar esse valor.



```python
class Valor:
    def __init__(self, data, progenitor=(), operador_mae="", rotulo=""):
        self.data = data
        self.progenitor = progenitor
        self.operador_mae = operador_mae
        self.rotulo = rotulo
        #Vamos criar uma propriedade de gradiente na nossa classe e atribuir um valor, ela não está nos argumentos porque queremos calcular o gradiente
        self.grad = 0

    def __repr__(self):
        return f"Valor(data={self.data})"

    def __add__(self, outro_valor):
        data = self.data + outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "+"
        saida = Valor(data, progenitor, operador_mae)
        return saida

    def __mul__(self, outro_valor):
        data = self.data * outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "*"
        saida = Valor(data, progenitor, operador_mae)
        return saida
```

Com essa alteração, agora o grafo computacional mostra o gradiente! Mas o valor do gradiente ainda está errado&#x2026;



```python
x1 = Valor(60, rotulo="x1")
x2 = Valor(24, rotulo="x2")
w1 = Valor(10, rotulo="w1")
w2 = Valor(5, rotulo="w2")
w3 = Valor(2, rotulo="w3")
b = Valor(7, rotulo="b")

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

plota_grafo(y)
```

### Calculando o gradiente local &ldquo;na mão&rdquo;



Vamos relembrar o cálculo do gradiente local e calcular os gradientes na mão! Para isso usaremos de noções básicas de derivada e da regra da cadeia.



```python
x1 = Valor(60, rotulo="x1")
x2 = Valor(24, rotulo="x2")
w1 = Valor(10, rotulo="w1")
w2 = Valor(5, rotulo="w2")
w3 = Valor(2, rotulo="w3")
b = Valor(7, rotulo="b")

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

# INICIA O BACKPROPAGATION
y.grad =  1 # dy/dy  (esse é o vértice folha!)

w3.grad = y.grad * k.data # dy/dw3
k.grad = y.grad * w3.data# dy/dk

b.grad =  k.grad * 1 # dy/db = dy/dk * dk/db
n.grad =  k.grad * 1 # dy/dn = dy/dk * dk/dn

s2.grad = n.grad * 1 # dy/ds2 = dy/dn * dn/s2
s1.grad = n.grad * 1 # dy/ds1 = dy/dn * dn/s1

x2.grad = s2.grad * w2.data # dy/dx2 = dy/ds2 * ds2/dx2
w2.grad = s2.grad * x2.data  # dy/dw2 = dy/ds2 * ds2/dw2

x1.grad = s1.grad * w1.data # dy/dx1 = dy/ds1 * ds1/dx1
w1.grad = s1.grad * x1.data # dy/dw1 = dy/ds1 * ds1/dw1

# FINALIZA O BACKPROPAGATION

plota_grafo(y)
```

### Treinando uma rede neural &ldquo;na mão&rdquo;



Agora que temos um grafo computacional com todos os gradientes locais calculados, podemos treinar a nossa rede neural manualmente.

Suponha, por um momento, que eu queira aumentar o valor de $y$ de saída da minha rede neural. Tem duas coisas que eu sei:

1.  eu posso apenas alterar os parâmetros internos da minha rede neural, isto é: $w_1$, $w_2$, $w_3$ e $b$. Os demais parâmetros eu não posso alterar pois ou são dados de entrada ($x_1$ e $x_2$) ou são parâmetros derivados de outros parâmetros ($k$, $n$, $s_1$, $s_2$ e $y$).

2.  o gradiente local indica o &ldquo;impacto&rdquo; que teremos no valor de $y$. Um gradiente local positivo me indica que ao aumentar o respectivo valor do vértice eu aumento o valor de $y$. Um gradiente local negativo me indica que ao aumentar o respectivo valor do vértice eu diminuo o valor de $y$.

Sabendo destes dois fatos, como que eu posso aumentar o valor de saída de $y$ apenas alterando os parâmetros internos da minha rede neural?



```python
taxa_de_aprendizado = 0.001

w1.data = w1.data + taxa_de_aprendizado * w1.grad
w2.data = w2.data + taxa_de_aprendizado * w2.grad
w3.data = w3.data + taxa_de_aprendizado * w3.grad
b.data = b.data + taxa_de_aprendizado * b.grad

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

y.grad =  1 # dy/dy  (esse é o vértice folha!)

w3.grad = y.grad * k.data # dy/dw3
k.grad = y.grad * w3.data# dy/dk

b.grad =  k.grad * 1 # dy/db = dy/dk * dk/db
n.grad =  k.grad * 1 # dy/dn = dy/dk * dk/dn

s2.grad = n.grad * 1 # dy/ds2 = dy/dn * dn/s2
s1.grad = n.grad * 1 # dy/ds1 = dy/dn * dn/s1

x2.grad = s2.grad * w2.data # dy/dx2 = dy/ds2 * ds2/dx2
w2.grad = s2.grad * x2.data  # dy/dw2 = dy/ds2 * ds2/dw2

x1.grad = s1.grad * w1.data # dy/dx1 = dy/ds1 * ds1/dx1
w1.grad = s1.grad * x1.data # dy/dw1 = dy/ds1 * ds1/dw1

plota_grafo(y)
```

### Cálculo automático do gradiente local



Calcular o gradiente local &ldquo;na mão&rdquo; não é a tarefa mais rápida nem a mais prazerosa que existe. Você certamente consegue pensar em uma, quem sabe até duas tarefas mais rápidas e prazerosas que essa. Tendo observado isso, pessoas refletiram se não seria possível de realizar esse cálculo de maneira automática para sobrar mais tempo na vida para se dedicar a outras tarefas. *Spoiler*: é possível!

Nossa classe `Valor` agora terá um método chamado `propagar`. Esse método irá, como o nome sugere, propagar o gradiente local de um certo vértice para seus progenitores.

Como padrão, o método `propagar` não irá fazer nada. Isso será o caso de vértices raiz que não têm progenitores (se não tem progenitor, então não tem o que propagar!).

Quando um vértice tem progenitores, daí teremos que sobrescrever o método `propagar` para que ele efetivamente propague os gradientes seguindo a regra da cadeia. Vamos fazer isso nos métodos `__add__` e `__mul__`.



```python
class Valor:
    def __init__(self, data, progenitor=(), operador_mae="", rotulo=""):
        self.data = data
        self.progenitor = progenitor
        self.operador_mae = operador_mae
        self.rotulo = rotulo
        self.grad = 0

    def __repr__(self):
        return f"Valor(data={self.data})"

    def __add__(self, outro_valor):
        data = self.data + outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "+"
        saida = Valor(data, progenitor, operador_mae)

        def propagar_adicao():
            self.grad = saida.grad * 1
            outro_valor.grad = saida.grad * 1

        saida.propagar = propagar_adicao

        return saida

    def __mul__(self, outro_valor):
        data = self.data * outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "*"
        saida = Valor(data, progenitor, operador_mae)

        def propagar_multiplicacao():
            self.grad = saida.grad * outro_valor.data
            outro_valor.grad = saida.grad * self.data

        saida.propagar = propagar_multiplicacao

        return saida

    def propagar(self):
        pass
```

Vamos testar!!



```python
x1 = Valor(60, rotulo="x1")
x2 = Valor(24, rotulo="x2")
w1 = Valor(10, rotulo="w1")
w2 = Valor(5, rotulo="w2")
w3 = Valor(2, rotulo="w3")
b = Valor(7, rotulo="b")

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

y.grad = 1  # precisamos iniciar o gradiente do vértice folha "na mão"
y.propagar()
k.propagar()
w3.propagar()
n.propagar()
b.propagar()
s1.propagar()
s2.propagar()
x1.propagar()
w1.propagar()
x2.propagar()
w2.propagar()

plota_grafo(y)
```

### Autograd



Temos, finalmente, todas as ferramentas para executar o que tanto queríamos: o algoritmo `backpropagation`. Para isso, nós tivemos que rodar o método `propagar` para todos os vértices do nosso grafo computacional. Seria tão mais interessante se esse processo todo fosse ainda mais automático!

Como você deve imaginar, podemos sim deixar esse processo ainda mais automático. Trata-se do algoritmo `autograd`! O último passo aqui é criar um método chamado `propagar_tudo` que cuida de fazer isso para nós.

O único detalhe que temos que nos atender é que não podemos rodar o método `propagar` em qualquer ordem dos vértices do grafo computacional. Apenas podemos rodar esse método quando sabemos o gradiente local do vértice filho! Uma estratégia para nos ajudar aqui é fazer uma [ordenação topológica](https://en.wikipedia.org/wiki/Topological_sorting). Ordenação topológica é quando você ordena os vértices de um grafo direcional de forma que todas as setas apontem para a direita. Nota: não podemos realizar ordenação topológica em grafos cíclicos. Aqui usaremos o código pronto feito pelo Andrej Karpathy que resolve esse problema. Veja abaixo de forma gráfica como funciona a ordenação topológica.

![img](https://i.imgur.com/Q3MA6dZ.png)



```python
class Valor:
    def __init__(self, data, progenitor=(), operador_mae="", rotulo=""):
        self.data = data
        self.progenitor = progenitor
        self.operador_mae = operador_mae
        self.rotulo = rotulo
        self.grad = 0

    def __repr__(self):
        return f"Valor(data={self.data})"

    def __add__(self, outro_valor):
        data = self.data + outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "+"
        saida = Valor(data, progenitor, operador_mae)

        def propagar():
            self.grad = saida.grad * 1
            outro_valor.grad = saida.grad * 1

        saida.propagar = propagar

        return saida

    def __mul__(self, outro_valor):
        data = self.data * outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "*"
        saida = Valor(data, progenitor, operador_mae)

        def propagar():
            self.grad = saida.grad * outro_valor.data
            outro_valor.grad = saida.grad * self.data

        saida.propagar = propagar

        return saida

    def propagar(self):
        pass

    def propagar_tudo(self):
        ordem_topologica = []
        visitados = set()

        def constroi_ordem_topologica(v):
            if v not in visitados:
                visitados.add(v)
                for progenitor in v.progenitor:
                    constroi_ordem_topologica(progenitor)
                ordem_topologica.append(v)

        constroi_ordem_topologica(self)

        self.grad = 1  # o gradiente do vértice folha deve ser 1

        for v in reversed(ordem_topologica):
            v.propagar()
```

Finalmente, depois de uma longa jornada, chegamos no nosso objetivo!



```python
x1 = Valor(60, rotulo="x1")
x2 = Valor(24, rotulo="x2")
w1 = Valor(10, rotulo="w1")
w2 = Valor(5, rotulo="w2")
w3 = Valor(2, rotulo="w3")
b = Valor(7, rotulo="b")

s1 = x1 * w1
s1.rotulo = "s1"

s2 = x2 * w2
s2.rotulo = "s2"

n = s1 + s2
n.rotulo = "n"

k = n + b
k.rotulo = "k"

y = k * w3
y.rotulo = "y"

y.propagar_tudo()

plota_grafo(y)
```

### Correção final para gradientes locais de funções de várias variáveis



Existe um problema com o cálculo do gradiente local da forma que estamos fazendo. Este bug aparece quando um mesmo vértice é usado para mais de uma operação no grafo computacional. Observe a situação abaixo. O gradiente local de $a$ deveria ser 4, porém o grafo mostra como sendo 1. Isto acontece pois a regra da cadeia para funções de várias variáveis nos diz que devemos somar os gradientes do vértice cada vez que chegamos nele mesmo, porém o nosso código não reflete isso. Isso não era um problema antes pois não estávamos usando o valor de um vértice duas ou mais vezes no grafo computacional.



```python
a = Valor(-2, rotulo="a")
b = Valor(3, rotulo="b")

c = a + b
d = a * b

e = c + d

c.rotulo = "c"
d.rotulo = "d"
e.rotulo = "e"

e.propagar_tudo()
plota_grafo(e)
```

Para a nossa sorte, é fácil de corrigir isso! Basta somar o gradiente com ele mesmo cada vez que ele for atualizado.



```python
class Valor:
    def __init__(self, data, progenitor=(), operador_mae="", rotulo=""):
        self.data = data
        self.progenitor = progenitor
        self.operador_mae = operador_mae
        self.rotulo = rotulo
        self.grad = 0

    def __repr__(self):
        return f"Valor(data={self.data})"

    def __add__(self, outro_valor):
        data = self.data + outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "+"
        saida = Valor(data, progenitor, operador_mae)

        def propagar():
            self.grad += saida.grad * 1
            outro_valor.grad += saida.grad * 1

        saida.propagar = propagar

        return saida

    def __mul__(self, outro_valor):
        data = self.data * outro_valor.data
        progenitor = (self, outro_valor)
        operador_mae = "*"
        saida = Valor(data, progenitor, operador_mae)

        def propagar():
            self.grad += saida.grad * outro_valor.data
            outro_valor.grad += saida.grad * self.data

        saida.propagar = propagar

        return saida

    def propagar(self):
        pass

    def propagar_tudo(self):
        ordem_topologica = []
        visitados = set()

        def constroi_ordem_topologica(v):
            if v not in visitados:
                visitados.add(v)
                for progenitor in v.progenitor:
                    constroi_ordem_topologica(progenitor)
                ordem_topologica.append(v)

        constroi_ordem_topologica(self)

        self.grad = 1  # o gradiente do vértice folha deve ser 1

        for v in reversed(ordem_topologica):
            v.propagar()
```

Com isso corrigimos esse bug!



```python
a = Valor(-2, rotulo="a")
b = Valor(3, rotulo="b")

c = a + b
d = a * b

e = c + d

c.rotulo = "c"
d.rotulo = "d"
e.rotulo = "e"

e.propagar_tudo()
plota_grafo(e)
```

## Conclusão


<p style="text-align: justify"> Neste experimento, abordamos o conceito de derivadas locais em redes neurais. O conceito de derivada local é de suma relevância para a aplicação do backpropagation de redes neurais, sendo o backpropagation, muito utilizado no processo de treinamento da rede, visto que ela ajusta os pesos e o viés  da rede neural de forma a otimiza-lá afim de reduzir o erro entre as saídas esperadas e as obtidas. E onde ligamos essa ideia com a postulação de gradientes? Pois bem, os gradientes locais são utilizados para o cálculo da contribuição de cada neurônio na rede neural de modo a determinar o seu erro total, ressalta-se que essas derivads são calculadas por meio da regra da cadeia, logo, calcula-se o gradiente de uma função composta em relação a cada uma das suas variáveis. </p>

<p style="text-align: justify"> Essencialmente, ele é utilizado na fase de backwars pass, já que o gradiente é calculado a partir do erro total em função da saída de rede e são propagados de volta para ajustar os pesos e o viés.  Esse cálculo ocorre por meio da determinação da derivadas parciais em cada camada da rede, representando como o erro total se propaga de volta pelos neurônios, por sua vez, a derivada local de um neurônio é calculada multiplicando o gradiente do neurônio seguinte pelo peso da conexão correspondente entre esses neurônio. Ademais, comenta-se que o cálculo de backwars pass é a segunda parte de funcionamento do backpropagation, já que inicialmente a rede inicia seus cálculos por meio do forward pass, sendo que as ativações de cada neurônio são calculadas à medida que as informações fluem da camada de entrada para a camada de saída.  </p>

<p style="text-align: justify"> Postulada a relevância do cálculo de gradientes locais, voltamos a analisar o experimento descrito no experimento, o qual aborda o cálculo do gradiente em seis etapas, nesta conclusão, iremos dissertar a respeito de todas.  </p>
<h4> Etapa 1: Preparo do cálculo do gradiente. </h4> 
<p style="text-align: justify">  A preparação do cálculo consiste na utilização de classes, constrói-se a classe "Valor" para a construção da propriedade "grad" para armazenar o valor de cada gradiente local porque cada valor númerico do grafo computacional deve ter um valor de gradiente local associado a ele. Inicialmente, estabelece-se o padrão de zero para a propriedade - o valor será alterado na Etapa 2.</p>

<h4> Etapa 2: Calculando o gradiente local "na mão". </h4>
<p style="text-align: justify"> Realizada a preparação, tem-se como abordagem o cálculo do gradiente "na mão", isto é, realizando o cálculo linha por linha por meio da regra da cadeia, sendo feito um backpass.</p>

<h4> Etapa 3: Treinando a rede neural "na mão". </h4>
<p style="text-align: justify"> Em seguida, realiza-se o treinamento da rede neural por meio do método de backpropagation. O treinamento consistiu na ideia de mudança do nosso valor de <b> y </b> - vértice final do grafo computacional; para isso, mudamos os valores internos da nossa rede <b> w1 </b>, <b> w2 </b>, <b> w3 </b> e <b> b</b> ao multiplica-lo por um valor denominado de taxa de aprendizagem. Ela possui esse nome, porque seria um valor aprendido que determina uma melhor resposta de saída da rede, afim de se diminuir o erro. </p>

<h4> Etapa 4: Calculando automaticamente o gradiente local. </h4>
<p style="text-align: justify"> Apesar da Etapa 1 ser uma excelente forma de se aprender a teória pos trás da obtenção do gradiente, ela é extremamente exaustiva. Assim, o questionamento que se perpetua é: "É possível automatizar esse processo?" e a resposta é afirmativa, sendo o gradiente obtido por meio de um incremento na classe "Valor" estabelecida na Etapa 1, sendo o cálculo realizado por um método chamado "propagar" que propaga o gradiente local de um certo vértice para seus progenitores. Esse método de propagar é utilizado quando um vértice tem progenitores, tendo que sobrescrever o método para que hava a propagação dos gradientes conforme a regra da cadeia. </p>

<h4> Etapa 5: Propagando o gradiente local automaticamente pela rede (AutoGRAD). </h4>
<p style="text-align: justify"> Neste momento, temos todas as ferramentas para realizamos o nosso backpropagation. Anteriormente realizamos um processo mais manual - rodamos o método propagar para todos os vértices do nosso grafo computacional, só que apesar desse ser um bom modelo para aprendermos o funcionamento de uma rede neural, na prática, realizamos um processo mais automático e aí que surge, o AutoGRAD que por meio da criação do método propagar que propaga tudo por nós, só tendo que nos atentar e executar uma propagação de ordenação topológica, isto é, devemos ordenar os vértices de um grafo direcional de forma que todas as setas apontem para a direita - cuidado, não podemos utilizar esse modelo em gráficos cíclicos, sendo necessário a alteração do código de ordenação. </p>

<h4> Etapa 6: Ajustando erros matemáticos. </h4>
<p style="text-align: justify"> Por fim, o experimento apresenta que o método utilizado pode ter erros matemáticos quando um mesmo vértice é utilizado para mais de uma operação no grafo computacional. Para isso, a correção utilizada no código é que ao ocorrer essa situação é somar o gradiente local que está dando erro, somando o gradiente com ele mesmo cada vez que ele for atualizado.</p>

<p style="text-align: justify"> Em síntese, esse é o funcionamento do experimento que computa os gradientes locais atomaticamente. </p>
